```python
import pandas as pd
```

```python
data = pd.read_csv('salaries.csv')
data.head()
```

```python
inputs = data.drop(['salary_more_then_100k','degree'], axis=1)
inputs
```

```python
inputs = data.drop('salary_more_then_100k', axis=1 )
target = data['salary_more_then_100k']
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_new'] = le_company.fit_transform(inputs['company'])
inputs['job_new'] = le_job.fit_transform(inputs['job'])
inputs['degree_new'] = le_degree.fit_transform(inputs['degree'])
```

```python
inputs
```

```python
inputs_new = inputs.drop(['company', 'job', 'degree'], axis=1)
```

```python
inputs_new
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(inputs_new, target)
```

```python
model.score(inputs_new, target)
```

```python
model.predict([[1,0,1]])
```
